```python
expense = []
budget = float(input("Enter budget for the month: "))

while True:
    print("\n--- Expense Tracker Menu ---")
    print("1. Add Expense")
    print("2. Monthly Summary")
    print("3. Category-wise Report")
    print("4. Budget Alert")
    print("5. Spending Insights")
    print("6. Exit")

    choice = input("Enter your choice: ")


    if choice == "1":
        date = input("Enter date (yyyy-mm-dd): ")
        cat = input("Enter category (Food, Travel, Shopping, etc.): ")
        amt = int(input("Enter the amount: "))

        expense.append({
            "date": date,
            "category": cat,
            "amount": amt
        })

        print("Expense added successfully.")


    elif choice == "2":
        total = 0
        for i in expense:
            total += i["amount"]
        print(f"Total Monthly Spending: {total}")


    elif choice == "3":
        cat_tot = {}

        for i in expense:
            if i["category"] in cat_tot:
                cat_tot[i["category"]] += i["amount"]
            else:
                cat_tot[i["category"]] = i["amount"]

        print("\nCategory-wise Spending:")
        for cat, amt in cat_tot.items():
            print(f"{cat}: {amt}")


    elif choice == "4":
        tot = 0
        for i in expense:
            tot += i["amount"]

        if tot > budget:
            print("âš  Monthly budget exceeded!")
        else:
            print("You are within your budget.")


    elif choice == "5":
        if len(expense) == 0:
            print("No expenses recorded.")
            continue


        day_total = {}
        for i in expense:
            if i["date"] in day_total:
                day_total[i["date"]] += i["amount"]
            else:
                day_total[i["date"]] = i["amount"]

        highest_day = ""
        highest_amount = 0
        for day in day_total:
            if day_total[day] > highest_amount:
                highest_amount = day_total[day]
                highest_day = day


        category_total = {}
        for i in expense:
            if i["category"] in category_total:
                category_total[i["category"]] += i["amount"]
            else:
                category_total[i["category"]] = i["amount"]

        dominant_category = ""
        max_spent = 0
        for cat in category_total:
            if category_total[cat] > max_spent:
                max_spent = category_total[cat]
                dominant_category = cat

        print(f"Highest spending day: {highest_day} (Amount: {highest_amount})")
        print(f"Dominant spending category: {dominant_category} (Amount: {max_spent})")


        avg_spending = 0
        for i in expense:
            avg_spending += i["amount"]
        avg_spending /= len(expense)

        print(f"Average Expense Amount: {avg_spending}")

        for i in expense:
            if i["amount"] > avg_spending:
                print(f"High spending detected on {i['date']} in {i['category']}")


    elif choice == "6":
        print("Exiting Expense Tracker. Goodbye!")
        break

    else:
        print("Invalid choice. Please try again.")
```

```python
courses = {}
students = {}
MAX_CREDITS = 20

while True:
    print("\n--- University Course Registration System ---")
    print("1. Add Course")
    print("2. Register Student")
    print("3. Enroll Student in Course")
    print("4. Student Summary Report")
    print("5. Course Summary Report")
    print("6. Exit")

    choice = input("Enter your choice: ")


    if choice == "1":
        course_id = input("Enter course ID: ")
        course_name = input("Enter course name: ")
        credits = int(input("Enter course credits: "))
        seats = int(input("Enter seat limit: "))

        courses[course_id] = {
            "name": course_name,
            "credits": credits,
            "seats": seats,
            "enrolled": 0
        }

        print("Course added successfully.")


    elif choice == "2":
        student_id = input("Enter student ID: ")
        student_name = input("Enter student name: ")

        students[student_id] = {
            "name": student_name,
            "courses": [],
            "total_credits": 0
        }

        print("Student registered successfully.")


    elif choice == "3":
        student_id = input("Enter student ID: ")
        course_id = input("Enter course ID: ")

        if student_id not in students:
            print("Student not found.")
        elif course_id not in courses:
            print("Course not found.")
        else:
            student = students[student_id]
            course = courses[course_id]

            if course_id in student["courses"]:
                print("Student already enrolled in this course.")
            elif course["enrolled"] >= course["seats"]:
                print("No seats available in this course.")
            elif student["total_credits"] + course["credits"] > MAX_CREDITS:
                print("Credit limit exceeded.")
            else:
                student["courses"].append(course_id)
                student["total_credits"] += course["credits"]
                course["enrolled"] += 1

                print("Student enrolled successfully.")


    elif choice == "4":
        for sid, info in students.items():
            print(f"\nStudent ID: {sid}")
            print(f"Name: {info['name']}")
            print(f"Enrolled Courses: {info['courses']}")
            print(f"Total Credits: {info['total_credits']}")


    elif choice == "5":
        for cid, info in courses.items():
            print(f"\nCourse ID: {cid}")
            print(f"Course Name: {info['name']}")
            print(f"Credits: {info['credits']}")
            print(f"Seats Filled: {info['enrolled']}/{info['seats']}")


    elif choice == "6":
        print("Exiting system. Goodbye!")
        break

    else:
        print("Invalid choice. Try again.")
```
